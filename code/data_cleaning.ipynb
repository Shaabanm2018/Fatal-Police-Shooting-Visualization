```python
import pandas as pd
```

```python
df=pd.read_csv('raw_data.csv')
df
```

```python
df_sankey = df.groupby(["state","race"])["name"].count().reset_index()
df_sankey.rename(columns={'state': 'abbr'}, inplace=True)
df_sankey.rename(columns={'name': 'count'}, inplace=True)
```

```python
states = {"AL":"Alabama","AK":"Alaska","AZ":"Arizona","AR":"Arkansas","CA":"California","CO":"Colorado","CT":"Connecticut","DE":"Delaware","DC":"WashingtonDC","FL":"Florida","GA":"Georgia","HI":"Hawaii","ID":"Idaho","IL":"Illinois","IN":"Indiana","IA":"Iowa","KS":"Kansas","KY":"Kentucky","LA":"Louisiana","ME":"Maine","MD":"Maryland","MA":"Massachusetts","MI":"Michigan","MN":"Minnesota","MS":"Mississippi","MO":"Missouri","MT":"Montana","NE":"Nebraska","NV":"Nevada","NH":"New Hampshire","NJ":"New Jersey","NM":"New Mexico","NY":"New York","NC":"North Carolina","ND":"North Dakota","OH":"Ohio","OK":"Oklahoma","OR":"Oregon","PA":"Pennsylvania","RI":"Rhode Island","SC":"South Carolina","SD":"South Dakota","TN":"Tennessee","TX":"Texas","UT":"Utah","VT":"Vermont","VA":"Virginia","WA":"Washington","WV":"West Virginia","WI":"Wisconsin","WY":"Wyoming"}
race = {"A":"Asian","W":"White","B":"Black","N":"NativeAmerican","O":"Other","H":"Hispanic"}
df_sankey["races"] = df_sankey["race"].map(race)
df_sankey["state"] = df_sankey["abbr"].map(states)
df_sankey.drop(['race','abbr'],axis=1)
df_sankey=df_sankey[['state','races','count']]
df_sankey.rename(columns={'races': 'target'}, inplace=True)
df_sankey['state'] = df_sankey['state'].str.upper() 
df_sankey.rename(columns={'state': 'source'}, inplace=True)
df_sankey.rename(columns={'count': 'value'}, inplace=True)
df_sankey
```

```python
df_sankey.to_csv('sankey.csv', encoding='utf-8', index=False)
```

```python
df_case_count = df.groupby(["state"])["name"].count().reset_index()
```

```python
df_case_count.rename(columns={'state': 'abbr'}, inplace=True)
```

```python
states = {"AL":"Alabama","AK":"Alaska","AZ":"Arizona","AR":"Arkansas","CA":"California","CO":"Colorado","CT":"Connecticut","DE":"Delaware","DC":"WashingtonDC","FL":"Florida","GA":"Georgia","HI":"Hawaii","ID":"Idaho","IL":"Illinois","IN":"Indiana","IA":"Iowa","KS":"Kansas","KY":"Kentucky","LA":"Louisiana","ME":"Maine","MD":"Maryland","MA":"Massachusetts","MI":"Michigan","MN":"Minnesota","MS":"Mississippi","MO":"Missouri","MT":"Montana","NE":"Nebraska","NV":"Nevada","NH":"New Hampshire","NJ":"New Jersey","NM":"New Mexico","NY":"New York","NC":"North Carolina","ND":"North Dakota","OH":"Ohio","OK":"Oklahoma","OR":"Oregon","PA":"Pennsylvania","RI":"Rhode Island","SC":"South Carolina","SD":"South Dakota","TN":"Tennessee","TX":"Texas","UT":"Utah","VT":"Vermont","VA":"Virginia","WA":"Washington","WV":"West Virginia","WI":"Wisconsin","WY":"Wyoming"}
df_case_count["state"] = df_case_count["abbr"].map(states)
df_case_count.rename(columns={'name': 'count'}, inplace=True)
df_case_count=df_case_count[['state','count']]
df_case_count['count_STR']=df_case_count['count']
df_case_count['count_STR'] = df_case_count['count_STR'].apply(str)
df_case_count['state'] = df_case_count['state'].str.upper() 
df_case_count
```

```python
df_case_count.to_csv('data.csv', encoding='utf-8', index=False)
```
